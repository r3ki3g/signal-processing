```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib nbagg
from IPython.display import Audio


sample_rate = 44100
frequency = 440  # Adjust this to your desired frequency in Hz

period = int(sample_rate / frequency)

pulse_wave = np.concatenate((np.ones(int(period / 2)), -np.ones(int(period / 2))))

duration = 1.0  # Adjust this to your desired duration in seconds
pulse_wave = np.tile(pulse_wave, int(duration * sample_rate // period + 1))
pulse_wave = pulse_wave[:int(duration * sample_rate)]

time = np.linspace(0, duration, len(pulse_wave))
plt.plot(time, pulse_wave)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Pulse Wave')
plt.show()

Audio(data=pulse_wave, rate=sample_rate,autoplay=1)
```
